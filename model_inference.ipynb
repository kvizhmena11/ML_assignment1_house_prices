```python
import pandas as pd
import numpy as np
import mlflow.sklearn
import joblib
import dagshub

dagshub.init(repo_owner='kvizhmena11', repo_name='ML_assignment1_house_prices', mlflow=True)
X_test = pd.read_pickle('X_test_cleaned.pkl')
test_ids = pd.read_pickle('test_ids.pkl')
X_test = X_test.fillna(0)
scaler = joblib.load('scaler.pkl')
X_test_scaled = scaler.transform(X_test)

model_uri = "models:/Champion_Ridge_Model/latest"
loaded_ridge_model = mlflow.sklearn.load_model(model_uri)

test_predictions_log = loaded_ridge_model.predict(X_test_scaled)
final_predictions = np.expm1(test_predictions_log)

submission = pd.DataFrame({
    'Id': test_ids,
    'SalePrice': final_predictions
})

submission.to_csv('ridge_submission.csv', index=False)
print("Submission saved successfully")
```
